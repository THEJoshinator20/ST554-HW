## Task 2: Summarize the Data

This data has many categorical variables and a few numeric. You should do the following:

### Categorical variables
* Create a one-way contingency table, a two-way contingency table, and a three-way contingency table
  * Interpret a number from each resulting table (that is, pick out a value produced and explain what that value means.)
* Create a conditional two-way table. That is, condition on one variable’s setting and create a two-way
table.

Do this using two different methods:
* Once, by subsetting the data (say with .loc) and then creating the two-way table
* Once, by creating a three-way table and subsetting it

```python
conttable1way = pd.crosstab(index = sdata["sex"], columns = 'count', margins=True)
print("\nOne Way Contingency Table\n")
print(conttable1way)
print("If we pick 187 that shows that there is a total of 187 boys in the datasets as a whole out of the 395.\nIt also shows that boys take up only ~47% of the total population of students.\n")
print('-' * 75)

print("\nTwo Way Contingency Table\n")
conttable2way = pd.crosstab(index = sdata["sex"], columns = sdata["Fjob"], margins=True)
print(conttable2way)
print("If we pick 106, that shows that 106 boys have a Mom that does a job that is not listed. \nAdditionally this means that boys who have mothers doing 'other' jobs out of all the boys is ~57%.\nAlso that out of the Mothers who are doing 'other' jobs ~49% of them have sons and not daughters.\n")
print('-' * 75)

print("\nThree Way Contingency Table\n")
conttable3way = pd.crosstab(index = sdata["sex"], columns = [sdata["Fjob"], sdata["Mjob"]], margins=True)
print(conttable3way)
print("If we pick 9, that shows that 9 female students have parents that both work as teachers. \nThis also means thatout of the total amount of girls ~4% of them have parents that are both teachers \nand out of the population of both parents being teachers ~47% have a daughter.\n")
print('-' * 75)

print("\nTwo Way Contingency Table Method 1\n")
condsdata = sdata.loc[sdata["Fjob"] == "other"]
condconttable2way = pd.crosstab(index = condsdata["sex"], columns = condsdata["Fjob"], margins=True)
print(condconttable2way)
print('-' * 75)

print("\nTwo Way Contingency Table Method 2\n")
condconttable3way = pd.crosstab(index = sdata["sex"], columns = [sdata["Fjob"], sdata["Mjob"]], margins=True)
condconttable3waysub = condconttable3way.loc[:, ("other", "teacher")]
print(condconttable3waysub)
```

### Numeric variables (and across groups)
The numeric variables are age, absences, and the three test grades variables (G1, G2, and G3).
* Find measures of center and spread for three of these variables (including G3 as one of them)
  * Repeat while subsetting the data by some grouping variable (say with .loc)
* Find measures of center and spread across a single grouping variable for three of these variables (including G3 as one of them)
* Find measures of center and spread across two grouping variables for three of these variables (including
G3 as one of them)
* Create a correlation matrix between all of the numeric variables

```python
age = sdata[["age", "absences", "G3"]].describe()
print("\nNumeric measures of Age, Absences, and Test Grades\n")
print(age)
print('-' * 75)

sdataage = sdata.loc[sdata["Pstatus"] == "A"]
agepstat = sdataage["age"].describe()
print("\nNumeric measures of Age while Pstatus is A\n")
print(agepstat)
print('-' * 75)

sdataabs = sdata.loc[sdata["famsize"] == "GT3"]
absencesfam = sdataabs["absences"].describe()
print("\nNumeric measures of Absences while famsize is GT3\n")
print(absencesfam)
print('-' * 75)

sdatag3 = sdata.loc[sdata["school"] == "GP"]
g3 = sdatag3["G3"].describe()
print("\nNumeric measures of Test Grades in 3rd Years while school is GP\n")
print(g3)
print('-' * 75)

#gpbyone = sdata.groupby("sex", group_keys = True)
gpbyone = sdata.groupby("sex")[["age", "absences", "G3"]].describe()
#gpbytwo = base.groupby(["sex", "Fjob"])
print("\nNumeric measures of Age, Absences, and Test Grades grouped by sex\n")
print(gpbyone)
print('-' * 75)

print("\nNumeric measures of Age, Absences, and Test Grades grouped by sex and Fjob\n")
gpbytwo = sdata.groupby(["sex", "Fjob"])[["age", "absences", "G3"]].describe()
print(gpbytwo)
print('-' * 75)

corrdata = sdata[["age", "absences", "G1", "G2", "G3"]].corr()
print("\nCorrelation Matrix\n")
print(corrdata)
```
