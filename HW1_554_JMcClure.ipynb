# Problem 1

# 1D Array

```python
axis0 = np.array([7, 2, 9, 10]) #first row
onedarray = np.array(axis0) #Not a needed function but useful in case 2d array is needed later

print(onedarray)
print(onedarray.shape)
print(onedarray.ndim)
```

# 2D Array

```python
axis0 = np.array([5.2, 3.0, 4.5]) #first row
axis1 = np.array([9.1, 0.1, 0.3]) #second row
twodarray = np.array([axis0, axis1]) # combine into full array

print(twodarray)
print(twodarray.shape)
print(twodarray.ndim)
```

# 3D Array

# Problem 2

```python
listy = [x for x in range(1,21)]
print(listy)
```

Method to remove 10

```python
listy.remove(10) #.remove is used to remove any value or string that is in the numbered position. You could choose to turn everything into a array and filter out a 10 but I didn't think that was what you were asking for.
print(listy)
```

Method to insert String 'threeve'

```python
listy.insert(0, 'threeve') #.insert is able to insert any string or number into any spot of your choosing. Techically number 1 is in position 0 but .insert shifts the 1 to the 1st position.
print(listy)
```

Method to repeat modified listy 3 times

```python
listy = listy * 3 #Everything in the list doesn't classifty as a number you can multiply the list variable by three to show it three times.
print(listy)
```
